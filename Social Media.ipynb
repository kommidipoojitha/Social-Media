```python

import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

tweets_data = pd.DataFrame({
    'tweet': [
        "Loving the new iPhone features! #Apple",
        "Can't believe I wasted money on this product. #disappointed",
        "Great experience at the concert last night. #awesome",
        "I'm feeling unsure about the new update. #confused",
        "Awful service at this restaurant. #terrible",
        "Enjoying the sunny weather today! #happy"
    ]
})

sid = SentimentIntensityAnalyzer()

def classify_sentiment(text):
    sentiment_score = sid.polarity_scores(text)
    if sentiment_score['compound'] >= 0.05:
        return 'positive'
    elif sentiment_score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

tweets_data['sentiment'] = tweets_data['tweet'].apply(classify_sentiment)

for index, row in tweets_data.iterrows():
    print(f'Tweet: {row["tweet"]}')
    print(f'Sentiment: {row["sentiment"]}')
    print('')

sentiment_counts = tweets_data['sentiment'].value_counts()
print("Sentiment Distribution:")
print(sentiment_counts)
```
